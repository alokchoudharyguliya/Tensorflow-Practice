```python
x_argmax=tf.constant([[2,30,20,3,6],[3,11,16,3,2]])
tf.math.argmax(x_argmax,axis=1) # axis=0, means across rows; axis=1, means across columns
# argmax, is for returning the indexes of the maximum number values
```

```python
tensor_two_d=tf.constant([[1,-2,0],[3,5,100],[1,5,6],[2,3,8]],dtype=tf.float16)
print(tensor_two_d.shape)
print(tf.math.reduce_std(tensor_two_d,axis=0,keepdims=True,name=None))
```

```python
tf.math.top_k(tensor_two_d,k=2)
```
